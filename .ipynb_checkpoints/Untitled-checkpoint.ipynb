```python
import pandas as pd
import matplotlib

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
salaries_df = pd.read_csv('./data/2018-salaries.csv')
batting_stats_df = pd.read_csv('./data/standard-stats-9-20-2018.csv')
```

```python
label_to_correlate = input()
```

```python
salaries_to_stats_df = pd.merge(salaries_df, batting_stats_df, left_on='Name', right_on='Name')
```

```python
salaries_to_stats_df
```

```python
scoped_df = salaries_to_stats_df.loc[:, ['Salary', label_to_correlate]]
scoped_df['Salary']=(scoped_df['Salary'].replace( '[\$,)]','', regex=True )
                   .replace( '[(]','-',   regex=True ).astype(float))
```

```python
X = scoped_df.iloc[:, :-1]
y = scoped_df.iloc[:, :1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0) 
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```
